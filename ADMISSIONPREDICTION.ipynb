```python
X=df.drop(['Chance of Admit '],axis=1)
```

```python
y = df['Chance of Admit ']
```

```python
from sklearn.model_selection import  train_test_split


train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print("setup complete")
```

```python
from sklearn import metrics
models = [['DecisionTree :',DecisionTreeRegressor()],
         ['Linear Regression :',LinearRegression()],
         ['KNeighbors :',KNeighborsRegressor(n_neighbors=3)]]
print("Results ..")

for name,model in models:
    model =model
    model.fit(train_x,train_y)
    predictions = model.predict(test_x)
    print(name,np.sqrt(mean_squared_error(test_y,predictions)))
    print(name,round((metrics.r2_score(test_y,predictions))*100))
```

```python
ks=10
mean_acc=np.zeros((ks-1))
for i in range(1,ks):
    knn = KNeighborsRegressor(n_neighbors=i).fit(train_x,train_y)
    y_hat = knn.predict(test_x)
    mean_acc[i-1]=metrics.r2_score(test_y,y_hat)
    print("Neighbor = %d"%i,"Accuracy ",np.round(mean_acc[i-1]*100),"%")

bac=mean_acc.max()*100
print("Best Acuracy",round(bac))
```
